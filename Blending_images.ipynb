```python
import numpy as np
import cv2
%matplotlib inline 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
#Started with OpenCV:Blending images
img_scene1=mpimg.imread('C:/Users/LENOVO/Desktop/Untitled Folder/scenery1.jfif')
resized=cv2.resize(img_scene1,(500,500))
plt.imshow(resized)
plt.title('Image1')
plt.show()
```

```python
img_scene2=mpimg.imread('C:/Users/LENOVO/Desktop/Untitled Folder/scenery2.jfif')
resized2 = cv2.resize(img_scene2, (500,500)) 

plt.imshow(resized2)
plt.title('Image2')

plt.show()
```

```python
blend = cv2.addWeighted(resized,0.5,resized2,0.6,0)
plt.imshow(blend)
plt.title('Blended Image')
plt.show()
```
